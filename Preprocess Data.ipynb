```python
import pandas as pd
import cv2
import os
import pickle
```

```python
csv_file = "/home/zzxxzz2569/VINAI_Chest_Xray/train_downsampled.csv"
df = pd.read_csv(csv_file)
print(df.head())
```

```python
train_folder = "/home/zzxxzz2569/VINAI_Chest_Xray/train/train"
idx=0
file_list = []
for file in os.listdir(train_folder):
    if file[0] != ".":
        print("Processing file {}- {}".format(idx, file))
        idx+=1
        
        # Find in CSV if there is any label for file and filter out class_id 14
        df_find = df[(df.image_id == file[:-4]) & (df.class_id!=14)]
        
        if len(df_find) > 0:
            raw_image = cv2.imread(os.path.join(train_folder, file),0)
            image_height, image_width = raw_image.shape[0], raw_image.shape[1]
            
            labels = []
            for index, row in df_find.iterrows():
                
                # Calculate coordinates and size by pixels
                box_width = row[6] - row[4]
                box_height = row[7] - row[5]
                box_center_x = (row[6] + row[4])/2
                box_center_y = (row[7] + row[5]) / 2
                
                # Normalization
                box_width_normalized = box_width / image_width
                box_height_normalized = box_height / image_height
                box_center_x_normalized = box_center_x / image_width
                box_center_y_normalized = box_center_y / image_height
                
                # Save those information to list
                labels.append([row[2], box_center_x_normalized, box_center_y_normalized, box_width_normalized, box_height_normalized])
                
            txt_file = file[:-4]+".txt"
            with open(os.path.join(train_folder, txt_file), 'w') as f:
                for label in labels:
                    f.write('{} {} {} {} {}\n'.format(label[0],label[1],label[2], label[3], label[4]))
            print("Labeled: ", txt_file)
            file_list.append(file)
            print("Number of labeled file: ", len(file_list))

# Save file_list to file pickle
with open('file_list.pkl', 'wb') as f:
    pickle.dump(file_list,f)
```

```python
import shutil
import os
import random
import pickle
```

```python
raw_folder = "/home/zzxxzz2569/VINAI_Chest_Xray/train/train"

with open('file_list.pkl', 'rb') as f:
    file_list = pickle.load(f)
    
total_files = len(file_list)
print("Total number of files: ", total_files)
```

```python
train_dir = "/home/zzxxzz2569/VINAI_Chest_Xray/yolo_data/images/train"
val_dir = "/home/zzxxzz2569/VINAI_Chest_Xray/yolo_data/images/test"

train_labels_dir = "/home/zzxxzz2569/VINAI_Chest_Xray/yolo_data/labels/train"
val_labels_dir = "/home/zzxxzz2569/VINAI_Chest_Xray/yolo_data/labels/test"
```

```python
if os.path.exists(train_dir) == False:
    os.makedirs(train_dir)
if os.path.exists(val_dir) == False:
    os.makedirs(val_dir)
if os.path.exists(train_labels_dir) == False:
    os.makedirs(train_labels_dir)
if os.path.exists(val_labels_dir) == False:
    os.makedirs(val_labels_dir)
```

```python
split_ratio = 0.2
total_val_files = int(split_ratio*total_files)
val_files = random.choices(file_list, k = total_val_files)
print("Number of val file: {}".format(len(val_files)))
```

```python
for file in  val_files:
    print("Validation file ", file)
    # Copy image
    shutil.copy(os.path.join(train_folder, file), os.path.join(val_dir,file))
    
    # Copy labels
    shutil.copy(os.path.join(train_folder, file[:-3]+'txt'), os.path.join(val_labels_dir, file[:-3]+'txt'))

for file in file_list:
    if file not in val_files:
        print("Train file ", file)
        # Copy image
        shutil.copy(os.path.join(train_folder, file), os.path.join(train_dir,file))

        # Copy labels
        shutil.copy(os.path.join(train_folder, file[:-3]+'txt'), os.path.join(train_labels_dir, file[:-3]+'txt'))
```
